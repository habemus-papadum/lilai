```python
import torch.nn as nn

class Inner(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Conv2d(3, 16, 3)
        self.bn = nn.BatchNorm2d(16)

class Outer(nn.Module):
    def __init__(self):
        super().__init__()
        self.inner = Inner()
        self.fc = nn.Linear(16, 10)

model = Outer()
for name, module in model.named_modules():
    print(f"{name!r:20} -> {module}")
```
